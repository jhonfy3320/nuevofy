```python
import matplotlib.pyplot as plt 

def generate_pie_chart():
    labels = ['A', 'B', 'C', 'D',]
    values = [12, 13, 14, 34]
    
    fig, ax = plt.subplots()
    ax.pie(values, labels=labels)
    plt.savefig('fig.png')
    plt.close()
    plt.show()
```

```python
import charts
def  run():
    charts.generate_pie_chart()
    
if __name__ == '__main__':
    run()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, label="Seno")
plt.xlabel("Tiempo")
plt.ylabel("Amplitud")
plt.title("Gráfico de una función seno")
plt.legend()
plt.show()
```
